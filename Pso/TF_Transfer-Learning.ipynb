```python
import os
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2(
    input_shape=(224,224, 3), alpha=1.0, include_top=False, weights='imagenet', classes=1000,
    classifier_activation='softmax')
```

```python
print(model.summary())
```

```python

base_inputs = model.layers[0].input
base_outputs = model.layers[-1].output
dense_outputs = layers.Dense(1000)(base_outputs)
normed_outputs = tf.norm(dense_outputs, ord='euclidean')
```

```python
new_model = keras.Model(inputs=base_inputs, outputs=normed_outputs)
new_model.compile(loss=keras.losses.CrossEntropyLoss
                 optimizer = keras.optimizers.Adam(),
                 metrics = ["accuracy"])
```

```python
new_model.save("OneDrive/Documents/GitHub/Job_Application_Projects/Pso/model", save_format='h5')
```
