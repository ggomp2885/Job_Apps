```python
import os
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2(
    input_shape=(224,224, 3), alpha=1.0, include_top=False, weights='imagenet', classes=1000,
    classifier_activation='softmax')
```

```python
print(model.summary())
```

```python

base_inputs = model.layers[0].input
base_outputs = model.layers[-1].output
dense_output = layers.Dense(1000)(base_outputs)
final_outputs = tf.norm(dense_output)


# print(dense_output)
#L2_norm_out_vector = np.linalg.norm(dense_output)
#print(L2_norm_out_vector)
```

```python
new_model = keras.Model(inputs=base_inputs, outputs=final_outputs)
print(new_model.summary())
```
